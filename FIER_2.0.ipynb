**AOI definition**
1.  Users first define a pin, line, and area with corresponded scale.
2.  The AOI will be the union of watershed that coincides with the defined pin, line, and area.

```python
def AOICreateByPoint(x,y,scale=6):
    ee_pin2 = ee.Geometry.Point(x,y)
    aoi2 =     ee.FeatureCollection("WWF/HydroSHEDS/v1/Basins/hybas_"+str(scale)).filterBounds(ee_pin2)
    Map = geemap.Map(center=(pin_lat_dict[pin_site], pin_lon_dict[pin_site]), zoom=8)
    url = 'https://mt1.google.com/vt/lyrs=m&x={x}&y={y}&z={z}&hl=en'
    Map.add_tile_layer(url, name='Google Maps', attribution='Google')
    Map.addLayer(ee_pin2)
    Map.addLayer(aoi2, {}, 'WWF HydroSHEDS (AOI)')
    return aoi2,Map
def AOICreateByLine(CoorList,scale=6):
    ee_pin2 = ee.Geometry.LineString(CoorList)
    aoi2 =     ee.FeatureCollection("WWF/HydroSHEDS/v1/Basins/hybas_"+str(scale)).filterBounds(ee_pin2)
    Map = geemap.Map(center=(pin_lat_dict[pin_site], pin_lon_dict[pin_site]), zoom=8)
    url = 'https://mt1.google.com/vt/lyrs=m&x={x}&y={y}&z={z}&hl=en'
    Map.add_tile_layer(url, name='Google Maps', attribution='Google')
    Map.addLayer(ee_pin2)
    Map.addLayer(aoi2, {}, 'WWF HydroSHEDS (AOI)')
    return aoi2,Map
def AOICreateByPol(CoorList,scale=6):
    ee_pin2 = ee.Geometry.Polygon(CoorList)
    aoi2 =     ee.FeatureCollection("WWF/HydroSHEDS/v1/Basins/hybas_"+str(scale)).filterBounds(ee_pin2)
    Map = geemap.Map(center=(pin_lat_dict[pin_site], pin_lon_dict[pin_site]), zoom=8)
    url = 'https://mt1.google.com/vt/lyrs=m&x={x}&y={y}&z={z}&hl=en'
    Map.add_tile_layer(url, name='Google Maps', attribution='Google')
    Map.addLayer(ee_pin2)
    Map.addLayer(aoi2, {}, 'WWF HydroSHEDS (AOI)')
    return aoi2,Map
```

```python
def folderCreater(pathIN,reNew=False):
    try:
        os.makedirs(pathIN)
    except:
        print('Folder: '+pathIN+'already exist')
        if reNew:
            shutil.rmtree(pathIN)
            os.makedirs(pathIN)
def int2Site(strIn):
    try:
        intIn=str(int(strIn)).zfill(len(strIn))
    except:
        intIn=0
    if intIn==strIn:
        return int(strIn)
    else:
        return strIn
def loadSmByDir(path,pinKey,tpcbool=True):
    RSM=xr.load_dataset(path+f'/RSM/{pinKey}_RSM.nc')
    RSMMean=xr.load_dataarray(path+f'/RSM/{pinKey}_RSM_MEAN.nc')
    tpcPath=glob.glob(path+'/hydro2rtpc_mdl/*.npy')
    tpcPath_poly=glob.glob(path+'/hydro2rtpc_mdl_poly/*.npy')
    print(tpcPath)
    #tcpMdl=[np.poly1d(np.load(i)) for i in tpcPath]
    tcpMdl=[np.load(i) for i in tpcPath]
    tcpMdl_poly=[np.load(i) for i in tpcPath_poly]

    if tpcbool==False:
        tcpMdl=tcpMdl_poly
    modeKey=[int(i.split('_tpc')[-1][:-4]) for i in tpcPath]
    siteKey=[int2Site(i.split('/site-')[-1].split('_tpc')[0]) for i in tpcPath]
    with open(path+f'/RSM/{pinKey}_RSM_r.txt','r') as f:
        reverseKey=[float(i)>0 for i in f.read().strip().split('\n')]
    return RSM,RSMMean,modeKey,reverseKey,siteKey,tcpMdl
```

**Create sigmoid Class**

```python
def sigmoid_F(x, A, h, slope, C):
    return 1 / (1 + np.exp ((x - h) * slope)) *  A + C
class  expFitHydrology:
    def __init__(self,dataIn):
        self.result=dataIn

    def __str__(self):
        return str(self.result)
    def __repr__(self):
        return str(self.result)
    def __call__(self, x_d):
        output=sigmoid_F(x_d,*self.result)
        return output
```

```python
def obs2SARLikeImage(doi,hist_q_stack,img_stackVV,RSSelect,RSMMean,modeKey,reverseKey,siteKey,tpcmdl):
    for i,modeN in enumerate(modeKey):
        site = siteKey[i]
        sm = RSSelect.spatial_modes.sel(mode=modeN)

        hydro_single = hist_q_stack.sel(site=site).dropna(dim='time')
        in_model =tpcmdl[i]
        indx_date = np.argwhere(hydro_single.time.values==doi)
        if indx_date.size==0:
            return False
        '''
        if reverseKey[i-1]==False:
            sm = sm*-1
        '''
        est_tpc = in_model(hydro_single.values[indx_date])
        #print(site)
        #print(hydro_single.values[indx_date])
        if i==0:
            syn_sar = sm*est_tpc
        else:
            syn_sar = syn_sar + sm*est_tpc
    syn_sar = syn_sar+RSMMean
    syn_sar.name = 'intensity'
    return syn_sar
```
